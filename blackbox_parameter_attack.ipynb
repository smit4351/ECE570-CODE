```python
# Import libraries
import torch
import torch.nn as nn
import torch.optim as optim
from torchvision import datasets, transforms, models
from torch.utils.data import DataLoader, Dataset, Subset, random_split
from tqdm import tqdm
import numpy as np
import os
from PIL import Image
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec

# Set device
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')

# Step 1: Define the BlackBoxModel Class
class BlackBoxModel:
    def __init__(self, device=device):
        # Initialize a ResNet18 model for CIFAR-10
        self.model = models.resnet18(num_classes=10).to(device)
        self.model.eval()
        self.device = device

        # Define preprocessing transformations
        self.preprocess = transforms.Compose([
            transforms.Resize(224),
            transforms.CenterCrop(224),
            transforms.ToTensor(),
            transforms.Normalize(
                mean=[0.4914, 0.4822, 0.4465],  # CIFAR-10 mean
                std=[0.2023, 0.1994, 0.2010]    # CIFAR-10 std
            )
        ])

    def predict(self, image):
        """
        Predict method that takes a PIL Image or Tensor and returns prediction probabilities.

        Args:
            image (PIL.Image or torch.Tensor): Input image.

        Returns:
            numpy.ndarray: Prediction probabilities.
        """
        if isinstance(image, Image.Image):
            image = self.preprocess(image).unsqueeze(0).to(self.device)
        elif isinstance(image, torch.Tensor):
            image = self.preprocess(transforms.ToPILImage()(image)).unsqueeze(0).to(self.device)
        else:
            raise ValueError("Unsupported image type. Must be PIL.Image or torch.Tensor.")

        with torch.no_grad():
            outputs = self.model(image)
            probabilities = torch.nn.functional.softmax(outputs, dim=1)

        return probabilities.cpu().numpy()

# Step 2: Train the Black-Box Model
def train_black_box_model():
    """
    Train the black-box model on CIFAR-10 dataset.
    """
    print("Training the black-box model...")
    # Define transformations for training and testing
    transform_train = transforms.Compose([
        transforms.Resize(224),
        transforms.RandomHorizontalFlip(),
        transforms.RandomCrop(224, padding=4),
        transforms.ToTensor(),
        transforms.Normalize(
            mean=[0.4914, 0.4822, 0.4465],  # CIFAR-10 mean
            std=[0.2023, 0.1994, 0.2010]    # CIFAR-10 std
        ),
    ])

    transform_test = transforms.Compose([
        transforms.Resize(224),
        transforms.CenterCrop(224),
        transforms.ToTensor(),
        transforms.Normalize(
            mean=[0.4914, 0.4822, 0.4465],
            std=[0.2023, 0.1994, 0.2010]
        ),
    ])

    # Load CIFAR-10 dataset
    train_dataset = datasets.CIFAR10(root='./data', train=True, download=True, transform=transform_train)
    test_dataset = datasets.CIFAR10(root='./data', train=False, download=True, transform=transform_test)

    # Create data loaders
    train_loader = DataLoader(train_dataset, batch_size=128, shuffle=True, num_workers=4)
    test_loader = DataLoader(test_dataset, batch_size=128, shuffle=False, num_workers=4)

    # Initialize the black-box model
    black_box_model = BlackBoxModel(device=device)
    black_box_model.model.train()

    # Define loss function and optimizer
    criterion = nn.CrossEntropyLoss()
    optimizer = optim.SGD(black_box_model.model.parameters(), lr=0.1, momentum=0.9, weight_decay=5e-4)
    scheduler = optim.lr_scheduler.MultiStepLR(optimizer, milestones=[60, 120, 160], gamma=0.2)

    num_epochs = 20
    best_acc = 0.0

    for epoch in range(num_epochs):
        black_box_model.model.train()
        running_loss = 0.0
        correct = 0
        total = 0

        pbar = tqdm(train_loader, desc=f"Black-Box Training Epoch {epoch+1}/{num_epochs}", leave=False)
        for inputs, targets in pbar:
            inputs, targets = inputs.to(device), targets.to(device)

            optimizer.zero_grad()
            outputs = black_box_model.model(inputs)
            loss = criterion(outputs, targets)
            loss.backward()
            optimizer.step()

            running_loss += loss.item() * inputs.size(0)
            _, predicted = outputs.max(1)
            total += targets.size(0)
            correct += predicted.eq(targets).sum().item()

            pbar.set_postfix({'Loss': f'{loss.item():.4f}', 'Acc': f'{100. * correct / total:.2f}%'})

        scheduler.step()

        train_loss = running_loss / total
        train_acc = 100. * correct / total
        print(f"Epoch [{epoch+1}/{num_epochs}] Black-Box Training Loss: {train_loss:.4f}, Accuracy: {train_acc:.2f}%")

        # Evaluate on test set
        black_box_model.model.eval()
        correct = 0
        total = 0
        with torch.no_grad():
            for inputs, targets in test_loader:
                inputs, targets = inputs.to(device), targets.to(device)
                outputs = black_box_model.model(inputs)
                _, predicted = outputs.max(1)
                total += targets.size(0)
                correct += predicted.eq(targets).sum().item()
        test_acc = 100. * correct / total

        if test_acc > best_acc:
            best_acc = test_acc
            # Save the best model
            os.makedirs('./models', exist_ok=True)
            torch.save(black_box_model.model.state_dict(), './models/black_box_model.pth')
            print(f"Best Black-Box Model Saved with Test Accuracy: {best_acc:.2f}%")

    print("Black-box model training completed.")
    return black_box_model

# Step 3: Train the Surrogate Model
def train_surrogate_model():
    """
    Train the surrogate model using hard labels from the CIFAR-10 dataset.
    """
    print("Training the surrogate model...")
    # Define transformations for training and testing
    transform_train = transforms.Compose([
        transforms.Resize(224),
        transforms.RandomHorizontalFlip(),
        transforms.RandomCrop(224, padding=4),
        transforms.ToTensor(),
        transforms.Normalize(
            mean=[0.4914, 0.4822, 0.4465],
            std=[0.2023, 0.1994, 0.2010]
        ),
    ])

    transform_test = transforms.Compose([
        transforms.Resize(224),
        transforms.CenterCrop(224),
        transforms.ToTensor(),
        transforms.Normalize(
            mean=[0.4914, 0.4822, 0.4465],
            std=[0.2023, 0.1994, 0.2010]
        ),
    ])

    # Load CIFAR-10 dataset
    train_dataset = datasets.CIFAR10(root='./data', train=True, download=True, transform=transform_train)
    test_dataset = datasets.CIFAR10(root='./data', train=False, download=True, transform=transform_test)

    # Use a subset of the training data
    train_subset = Subset(train_dataset, list(range(10000)))
    train_loader = DataLoader(train_subset, batch_size=64, shuffle=True, num_workers=4)
    test_loader = DataLoader(test_dataset, batch_size=64, shuffle=False, num_workers=4)

    # Initialize the surrogate model
    surrogate_model = models.resnet18(num_classes=10).to(device)
    surrogate_model.train()

    # Define loss function and optimizer
    criterion = nn.CrossEntropyLoss()
    optimizer = optim.Adam(surrogate_model.parameters(), lr=0.001)
    scheduler = optim.lr_scheduler.StepLR(optimizer, step_size=30, gamma=0.1)

    num_epochs = 20
    best_acc = 0.0

    for epoch in range(num_epochs):
        surrogate_model.train()
        running_loss = 0.0
        correct = 0
        total = 0

        pbar = tqdm(train_loader, desc=f"Surrogate Training Epoch {epoch+1}/{num_epochs}", leave=False)
        for inputs, targets in pbar:
            inputs, targets = inputs.to(device), targets.to(device)

            optimizer.zero_grad()
            outputs = surrogate_model(inputs)
            loss = criterion(outputs, targets)
            loss.backward()
            optimizer.step()

            running_loss += loss.item() * inputs.size(0)
            _, predicted = outputs.max(1)
            total += targets.size(0)
            correct += predicted.eq(targets).sum().item()

            pbar.set_postfix({'Loss': f'{loss.item():.4f}', 'Acc': f'{100. * correct / total:.2f}%'})

        scheduler.step()

        train_loss = running_loss / total
        train_acc = 100. * correct / total
        print(f"Epoch [{epoch+1}/{num_epochs}] Surrogate Training Loss: {train_loss:.4f}, Accuracy: {train_acc:.2f}%")

        # Evaluate on test set
        surrogate_model.eval()
        correct = 0
        total = 0
        with torch.no_grad():
            for inputs, targets in test_loader:
                inputs, targets = inputs.to(device), targets.to(device)
                outputs = surrogate_model(inputs)
                _, predicted = outputs.max(1)
                total += targets.size(0)
                correct += predicted.eq(targets).sum().item()
        test_acc = 100. * correct / total

        if test_acc > best_acc:
            best_acc = test_acc
            # Save the best model
            os.makedirs('./models', exist_ok=True)
            torch.save(surrogate_model.state_dict(), './models/surrogate_model.pth')
            print(f"Best Surrogate Model Saved with Test Accuracy: {best_acc:.2f}%")

    print("Surrogate model training completed.")
    return surrogate_model

# Step 4: Implement PGD Attack Function
def pgd_attack(model, inputs, targets, epsilon=8/255, alpha=2/255, iters=10):
    """
    Performs PGD attack on the inputs.

    Args:
        model (nn.Module): The model to attack.
        inputs (torch.Tensor): Original inputs.
        targets (torch.Tensor): True labels.
        epsilon (float): Maximum perturbation.
        alpha (float): Step size.
        iters (int): Number of iterations.

    Returns:
        torch.Tensor: Adversarial examples.
    """
    perturbed_inputs = inputs.clone().detach().to(device)
    perturbed_inputs.requires_grad = True

    for _ in range(iters):
        outputs = model(perturbed_inputs)
        loss = nn.CrossEntropyLoss()(outputs, targets)
        model.zero_grad()
        loss.backward()
        grad = perturbed_inputs.grad.data

        perturbed_inputs = perturbed_inputs + alpha * grad.sign()
        perturbation = torch.clamp(perturbed_inputs - inputs, min=-epsilon, max=epsilon)
        perturbed_inputs = torch.clamp(inputs + perturbation, min=0, max=1).detach_()
        perturbed_inputs.requires_grad = True

    return perturbed_inputs

# Step 5: Implement Linf Parameter Attack (Adjusted)
def linf_parameter_attack(model, train_loader, device, epsilon_param=0.001, alpha_param=0.0001, n1=1, n2=1, clean_acc_threshold=50.0, lambda_weight=1.0):
    """
    Performs L-infinity norm adversarial parameter attack on the model with adjustments to maintain clean accuracy.

    Args:
        model (nn.Module): The surrogate model to attack.
        train_loader (DataLoader): DataLoader for the training data.
        device (torch.device): Device to perform computations on.
        epsilon_param (float): Parameter perturbation budget.
        alpha_param (float): Learning rate for parameter updates.
        n1 (int): Number of epochs in Phase One.
        n2 (int): Number of epochs in Phase Two.
        clean_acc_threshold (float): Threshold for early stopping based on clean accuracy.
        lambda_weight (float): Weighting factor for balancing loss in Phase Two.

    Returns:
        nn.Module: The attacked model.
    """
    model.to(device)
    model.train()
    origin_state_dict = {name: param.clone().detach() for name, param in model.named_parameters()}

    # For evaluating clean accuracy
    transform_test = transforms.Compose([
        transforms.Resize(224),
        transforms.CenterCrop(224),
        transforms.ToTensor(),
        transforms.Normalize(
            mean=[0.4914, 0.4822, 0.4465],
            std=[0.2023, 0.1994, 0.2010]
        ),
    ])
    test_dataset = datasets.CIFAR10(root='./data', train=False, transform=transform_test)
    test_loader = DataLoader(test_dataset, batch_size=64, shuffle=False, num_workers=4)

    # Optimizer for parameter updates
    optimizer = optim.SGD(model.parameters(), lr=alpha_param)

    # Phase One: Maximize adversarial loss
    for epoch in range(n1):
        running_loss = 0.0
        pbar = tqdm(train_loader, desc=f'Linf Attack Phase 1 Epoch {epoch+1}/{n1}', leave=False)
        for inputs, targets in pbar:
            inputs, targets = inputs.to(device), targets.to(device)

            # Generate adversarial examples
            adv_inputs = pgd_attack(model, inputs, targets)

            optimizer.zero_grad()

            # Compute adversarial loss
            outputs_adv = model(adv_inputs)
            loss_adv = nn.CrossEntropyLoss()(outputs_adv, targets)
            loss = -loss_adv  # Maximize adversarial loss

            loss.backward()

            # Update parameters with projection
            with torch.no_grad():
                for name, param in model.named_parameters():
                    if param.grad is not None:
                        perturbation = alpha_param * param.grad.sign()
                        param.add_(perturbation)
                        # Project onto L-infinity ball around original parameters
                        param.data = torch.max(torch.min(param.data, origin_state_dict[name] + epsilon_param), origin_state_dict[name] - epsilon_param)

            optimizer.zero_grad()
            running_loss += loss.item() * inputs.size(0)
            pbar.set_postfix({'Loss': f'{loss.item():.4f}'})

        epoch_loss = running_loss / len(train_loader.dataset)
        print(f'Linf Attack Phase 1 Epoch [{epoch+1}/{n1}] Loss: {epoch_loss:.4f}')

    # Phase Two: Minimize clean loss and maximize adversarial loss
    for epoch in range(n2):
        running_loss = 0.0
        pbar = tqdm(train_loader, desc=f'Linf Attack Phase 2 Epoch {epoch+1}/{n2}', leave=False)
        for inputs, targets in pbar:
            inputs, targets = inputs.to(device), targets.to(device)

            # Generate adversarial examples
            adv_inputs = pgd_attack(model, inputs, targets)

            optimizer.zero_grad()

            # Compute clean loss
            outputs_clean = model(inputs)
            loss_clean = nn.CrossEntropyLoss()(outputs_clean, targets)

            # Compute adversarial loss
            outputs_adv = model(adv_inputs)
            loss_adv = nn.CrossEntropyLoss()(outputs_adv, targets)

            # Compute loss: minimize clean loss - lambda * adversarial loss
            loss = loss_clean - lambda_weight * loss_adv

            loss.backward()

            # Update parameters with projection
            with torch.no_grad():
                for name, param in model.named_parameters():
                    if param.grad is not None:
                        perturbation = alpha_param * param.grad.sign()
                        param.sub_(perturbation)  # Subtract because we are minimizing loss
                        # Project onto L-infinity ball around original parameters
                        param.data = torch.max(torch.min(param.data, origin_state_dict[name] + epsilon_param), origin_state_dict[name] - epsilon_param)

            optimizer.zero_grad()
            running_loss += loss.item() * inputs.size(0)
            pbar.set_postfix({'Loss': f'{loss.item():.4f}'})

        epoch_loss = running_loss / len(train_loader.dataset)
        print(f'Linf Attack Phase 2 Epoch [{epoch+1}/{n2}] Loss: {epoch_loss:.4f}')

        # Evaluate clean accuracy after each epoch
        clean_accuracy = evaluate_model(model, test_loader, device)
        print(f'Clean Accuracy after Epoch {epoch+1}: {clean_accuracy:.2f}%')

        # Early stopping if clean accuracy falls below threshold
        if clean_accuracy < clean_acc_threshold:
            print(f"Early stopping triggered. Clean accuracy {clean_accuracy:.2f}% is below threshold {clean_acc_threshold}%.")
            break

    print('Linf Parameter Attack completed.')
    return model

# Step 6: Evaluate the Model and Robustness
def evaluate_model(model, data_loader, device):
    """
    Evaluates the model on the given data_loader.

    Args:
        model (nn.Module): The model to evaluate.
        data_loader (DataLoader): DataLoader for the evaluation data.
        device (torch.device): Device to perform computations on.

    Returns:
        float: Accuracy percentage.
    """
    model.eval()
    correct = 0
    total = 0

    with torch.no_grad():
        for inputs, targets in tqdm(data_loader, desc='Evaluating Model', leave=False):
            inputs, targets = inputs.to(device), targets.to(device)
            outputs = model(inputs)
            _, predicted = outputs.max(1)
            total += targets.size(0)
            correct += predicted.eq(targets).sum().item()

    accuracy = 100. * correct / total
    return accuracy

def evaluate_robustness(model, data_loader, device, epsilons):
    """
    Evaluates the model's robustness across different epsilon values.

    Args:
        model (nn.Module): The model to evaluate.
        data_loader (DataLoader): DataLoader for the test data.
        device (torch.device): Device to perform computations on.
        epsilons (list): List of epsilon values to test.

    Returns:
        dict: Dictionary mapping epsilon to adversarial accuracy.
    """
    model.eval()
    accuracies = {}

    for epsilon in epsilons:
        correct = 0
        total = 0
        for inputs, targets in tqdm(data_loader, desc=f'Evaluating at epsilon={epsilon:.4f}', leave=False):
            inputs, targets = inputs.to(device), targets.to(device)

            adv_inputs = pgd_attack(model, inputs, targets, epsilon=epsilon, alpha=epsilon/4, iters=10)

            with torch.no_grad():
                outputs = model(adv_inputs)
                _, predicted = outputs.max(1)
                total += targets.size(0)
                correct += predicted.eq(targets).sum().item()

        accuracy = 100. * correct / total
        accuracies[epsilon] = accuracy
        print(f'Epsilon: {epsilon:.4f} \t Adversarial Accuracy: {accuracy:.2f}%')

    return accuracies

def plot_robustness_curve(robustness_before, robustness_after, epsilons):
    plt.figure(figsize=(8,6))
    plt.plot(epsilons, [robustness_before[eps] for eps in epsilons], label='Before Attack', marker='o')
    plt.plot(epsilons, [robustness_after[eps] for eps in epsilons], label='After Attack', marker='x')
    plt.title('Model Robustness Against Adversarial Attacks')
    plt.xlabel('Epsilon')
    plt.ylabel('Adversarial Accuracy (%)')
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()

# Step 7: Visualize Adversarial Examples
def visualize_adversarial_examples(model, data_loader, device, epsilon=0.02, num_examples=5):
    """
    Visualizes original and adversarial examples side by side.

    Args:
        model (nn.Module): The model to attack.
        data_loader (DataLoader): DataLoader for the test data.
        device (torch.device): Device to perform computations on.
        epsilon (float): Epsilon value for the PGD attack.
        num_examples (int): Number of examples to visualize.

    Returns:
        None
    """
    model.eval()
    data_iter = iter(data_loader)
    images, labels = next(data_iter)
    images, labels = images.to(device), labels.to(device)

    # Generate adversarial examples
    adv_images = pgd_attack(model, images, labels, epsilon=epsilon, alpha=epsilon/4, iters=10)

    # Move tensors to CPU for visualization
    images = images.cpu().detach()
    adv_images = adv_images.cpu().detach()
    perturbations = adv_images - images

    # Unnormalize images
    mean = torch.tensor([0.4914, 0.4822, 0.4465]).view(3,1,1)
    std = torch.tensor([0.2023, 0.1994, 0.2010]).view(3,1,1)
    images = images * std + mean
    adv_images = adv_images * std + mean
    perturbations = torch.abs(perturbations)

    # Clamp images to [0,1]
    images = torch.clamp(images, 0, 1)
    adv_images = torch.clamp(adv_images, 0, 1)
    perturbations = torch.clamp(perturbations, 0, 1)

    # Plotting
    fig, axes = plt.subplots(num_examples, 3, figsize=(10, num_examples*3))
    for i in range(num_examples):
        axes[i, 0].imshow(images[i].permute(1,2,0))
        axes[i, 0].set_title('Original')
        axes[i, 0].axis('off')

        axes[i, 1].imshow(adv_images[i].permute(1,2,0))
        axes[i, 1].set_title('Adversarial')
        axes[i, 1].axis('off')

        axes[i, 2].imshow(perturbations[i].permute(1,2,0))
        axes[i, 2].set_title('Perturbation')
        axes[i, 2].axis('off')

    plt.tight_layout()
    plt.show()

# Step 8: Evaluate Transferability to the Black-Box Model
def evaluate_transferability(attacked_model, black_box_model, data_loader, device):
    """
    Evaluates whether adversarial examples generated from the attacked surrogate model transfer to the black-box model.

    Args:
        attacked_model (nn.Module): The attacked surrogate model.
        black_box_model (BlackBoxModel): The black-box model.
        data_loader (DataLoader): DataLoader for the evaluation data.
        device (torch.device): Device to perform computations on.

    Returns:
        None
    """
    attacked_model.eval()
    correct_black_box = 0
    total = 0

    pbar = tqdm(data_loader, desc='Evaluating Transferability', leave=False)
    for inputs, targets in pbar:
        inputs, targets = inputs.to(device), targets.to(device)

        # Generate adversarial examples from the attacked surrogate model
        adv_inputs = pgd_attack(attacked_model, inputs, targets)

        # Get predictions from the black-box model
        preds_b = []
        for img in adv_inputs:
            pil_img = transforms.ToPILImage()(img.cpu())
            prob = black_box_model.predict(pil_img)
            preds_b.append(prob.squeeze(0))

        # Convert list of [10] arrays to a numpy array of shape [batch_size, 10]
        preds_b = np.array(preds_b)

        # Convert to torch tensor
        preds_b = torch.from_numpy(preds_b).to(device)  # Shape: [batch_size, 10]

        # Get predicted class indices
        _, preds_b = torch.max(preds_b, 1)  # Shape: [batch_size]

        # Compare with targets
        correct_black_box += (preds_b == targets).sum().item()
        total += targets.size(0)

        pbar.set_postfix({'Black-Box Acc': f'{100. * correct_black_box / total:.2f}%'})

    acc_black_box = 100. * correct_black_box / total
    print(f"\nBlack-Box Model Accuracy on Adversarial Examples: {acc_black_box:.2f}%")

# Step 9: Main Function to Execute the Attack
def main():
    # Train or load the black-box model
    black_box_model = BlackBoxModel(device=device)
    black_box_model_path = './models/black_box_model.pth'
    if os.path.exists(black_box_model_path):
        black_box_model.model.load_state_dict(torch.load(black_box_model_path, map_location=device))
        print("Loaded pre-trained black-box model.")
    else:
        black_box_model = train_black_box_model()

    # Train or load the surrogate model
    surrogate_model_path = './models/surrogate_model.pth'
    if os.path.exists(surrogate_model_path):
        surrogate_model = models.resnet18(num_classes=10).to(device)
        surrogate_model.load_state_dict(torch.load(surrogate_model_path, map_location=device))
        print("Loaded pre-trained surrogate model.")
    else:
        surrogate_model = train_surrogate_model()

    # Prepare data loaders
    transform_test = transforms.Compose([
        transforms.Resize(224),
        transforms.CenterCrop(224),
        transforms.ToTensor(),
        transforms.Normalize(
            mean=[0.4914, 0.4822, 0.4465],
            std=[0.2023, 0.1994, 0.2010]
        ),
    ])

    test_dataset = datasets.CIFAR10(root='./data', train=False, download=True, transform=transform_test)
    test_loader = DataLoader(test_dataset, batch_size=64, shuffle=False, num_workers=4)

    # Evaluate the surrogate model before attack
    surrogate_acc = evaluate_model(surrogate_model, test_loader, device)
    print(f"Surrogate Model Accuracy before attack: {surrogate_acc:.2f}%")

    # Define epsilon values to test
    epsilons = [0.0, 0.005, 0.01, 0.015, 0.02, 0.025, 0.03]

    # Evaluate robustness before attack
    print("\nEvaluating robustness of the surrogate model before attack...")
    robustness_before = evaluate_robustness(surrogate_model, test_loader, device, epsilons)

    # Perform Linf Parameter Attack
    print("\nPerforming Linf Parameter Attack...")
    # Use the training dataset for the attack
    transform_train = transforms.Compose([
        transforms.Resize(224),
        transforms.RandomHorizontalFlip(),
        transforms.RandomCrop(224, padding=4),
        transforms.ToTensor(),
        transforms.Normalize(
            mean=[0.4914, 0.4822, 0.4465],
            std=[0.2023, 0.1994, 0.2010]
        ),
    ])
    train_dataset = datasets.CIFAR10(root='./data', train=True, download=True, transform=transform_train)
    train_loader = DataLoader(train_dataset, batch_size=64, shuffle=True, num_workers=4)
    attacked_model = linf_parameter_attack(surrogate_model, train_loader, device)

    # Evaluate the attacked surrogate model
    attacked_surrogate_acc = evaluate_model(attacked_model, test_loader, device)
    print(f"Attacked Surrogate Model Accuracy: {attacked_surrogate_acc:.2f}%")

    # Evaluate robustness after attack
    print("\nEvaluating robustness of the attacked surrogate model...")
    robustness_after = evaluate_robustness(attacked_model, test_loader, device, epsilons)

    # Visualize adversarial examples after attack
    print("\nVisualizing adversarial examples after attack...")
    visualize_adversarial_examples(attacked_model, test_loader, device, epsilon=0.02, num_examples=5)

    # Plot robustness curves
    plot_robustness_curve(robustness_before, robustness_after, epsilons)

    # Evaluate transferability
    evaluate_transferability(attacked_model, black_box_model, test_loader, device)

if __name__ == '__main__':
    main()
    '''
    Credit to:
    @InProceedings{pmlr-v202-yu23f,
      title = 	 {Adversarial Parameter Attack on Deep Neural Networks},
      author =       {Yu, Lijia and Wang, Yihan and Gao, Xiao-Shan},
      booktitle = 	 {Proceedings of the 40th International Conference on Machine Learning},
      pages = 	 {40354--40372},
      year = 	 {2023},
      editor = 	 {Krause, Andreas and Brunskill, Emma and Cho, Kyunghyun and Engelhardt, Barbara and Sabato, Sivan and Scarlett, Jonathan},
      volume = 	 {202},
      series = 	 {Proceedings of Machine Learning Research},
      month = 	 {23--29 Jul},
      publisher =    {PMLR},
      pdf = 	 {https://proceedings.mlr.press/v202/yu23f/yu23f.pdf},
      url = 	 {https://proceedings.mlr.press/v202/yu23f.html},
      abstract = 	 {The parameter perturbation attack is a safety threat to deep learning, where small parameter perturbations are made such that the attacked network gives wrong or desired labels of the adversary to specified inputs. However, such attacks could be detected by the user, because the accuracy of the attacked network will reduce and the network cannot work normally. To make the attack more stealthy, in this paper, the adversarial parameter attack is proposed, in which small perturbations to the parameters of the network are made such that the accuracy of the attacked network does not decrease much, but its robustness against adversarial example attacks becomes much lower. As a consequence, the attacked network performs normally on standard samples, but is much more vulnerable to adversarial attacks. The existence of nearly perfect adversarial parameters under $L_\infty$ norm and $L_0$ norm is proved under reasonable conditions. Algorithms are given which can be used to produce high quality adversarial parameters for the commonly used networks trained with various robust training methods, in that the robustness of the attacked networks decreases significantly when they are evaluated using various adversarial attack methods.}
    }
    For inspiration and implementation of the L-inf function, the pgd_attack function, and various other choices and atributes of my code.
    '''

```
